```
from transformers import  FSMTTokenizer, FSMTForConditionalGeneration
```

```
source = 'facebook/wmt19-ru-en'
tokenizer = FSMTTokenizer.from_pretrained(source)
model = FSMTForConditionalGeneration.from_pretrained(source)
```

```
def rus2en(sent):
    res = model.generate(tokenizer.encode(sent, return_tensors='pt'))
    decoded = tokenizer.decode(res[0], skip_special_tokens=True)
    return decoded
```

```
path_to_test = '/content/test.ru.txt'
path_to_eval = '/content/eval-ru-100.txt'
```

```
def read_data(input_path):
    sentences = []
    with open(input_path, 'r') as test:
        data = test.read()
        lines = data.split('\n')
    return lines
eval_data = read_data(path_to_eval)
eval_data
```

```
def write_to_file(output, out_path='answer.txt'):
    with open(out_path, 'w') as out_file:
        out_file.write('\n'.join(output))
```
